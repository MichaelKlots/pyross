```python
import numpy as np
import pyross
import matplotlib.pyplot as plt
from matplotlib import rc; rc('text', usetex=True)
```

## Age-structures

```python
my_data = np.genfromtxt('../data/age_structures/China-2019.csv', delimiter=',', skip_header=1)
aM, aF = my_data[:, 1], my_data[:, 2]

print(aM)
chinaM = 100*aM/np.sum(2*aM)
chinaF = 100*aF/np.sum(2*aF)
```

```python
my_data = np.genfromtxt('../data/age_structures/Denmark-2019.csv', delimiter=',', skip_header=1)
aM, aF = my_data[:, 1], my_data[:, 2]

print(aM)
denM = 100*aM/np.sum(2*aM)
denF = 100*aF/np.sum(2*aF)
```

```python
my_data = np.genfromtxt('../data/age_structures/India-2019.csv', delimiter=',', skip_header=1)
aM, aF = my_data[:, 1], my_data[:, 2]

print(aM)
indiaM = 100*aM/np.sum(2*aM)
indiaF = 100*aF/np.sum(2*aF)
```

```python
my_data = np.genfromtxt('../data/age_structures/Italy-2019.csv', delimiter=',', skip_header=1)
aM, aF = my_data[:, 1], my_data[:, 2]

print(aM)
italyM = 100*aM/np.sum(2*aM)
italyF = 100*aF/np.sum(2*aF)
```

## Contact structures

```python
chCH, chCW, chCS, chCO = pyross.contactMatrix.China()
deCH, deCW, deCS, deCO = pyross.contactMatrix.Denmark()
inCH, inCW, inCS, inCO = pyross.contactMatrix.India()
itCH, itCW, itCS, itCO = pyross.contactMatrix.Italy()
```
