```python
import numpy as np
import pyross, odespy
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc; rc('text', usetex=True)
```

```python
my_data = np.genfromtxt('../../data/age_structures/India-2019.csv', delimiter=',', skip_header=1)
aM, aF = my_data[:, 1], my_data[:, 2]

aM = 100*aM/np.sum(aM)
aF = 100*aF/np.sum(aF)
```

```python
fig = plt.figure(num=None, figsize=(8, 12), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 22})

sAge = np.arange((np.size(aM)))

plt.barh(sAge, aM, color="#348ABD", label='Male', alpha=0.6)
plt.barh(sAge, -aF,color='#A60628', label='Female', alpha=0.6)

plt.legend(fontsize=26); plt.grid() 
```
